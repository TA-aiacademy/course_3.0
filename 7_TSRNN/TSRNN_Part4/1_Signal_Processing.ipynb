````python
import librosa
import IPython.display as idp # æ’­éŸ³å·¥å…·
import numpy as np # è¼”åŠ©é‹ç®—
import matplotlib.pyplot as plt # è¼”åŠ©ç•«åœ–
from plotly import express as px
````

````python
def plot_signal(signal,sr, start=0, end=None, labels=None,title=None):
    
    ## Visualizes signal data
    ## Args:
    #  signal (list of array of int) - æ™‚é–“é»å°æ‡‰çš„è¨Šè™Ÿï¼Œåˆ—è¡¨å…§æ™‚é–“åºåˆ—æ•¸é‡ç‚ºDï¼Œæ¯ç­†è³‡æ–™é•·åº¦ç‚ºTï¼Œè‹¥éç‚ºåˆ—è¡¨å‰‡è½‰ç‚ºåˆ—è¡¨
    #  start (int) - é–‹å§‹çš„è³‡æ–™åº(ç¬¬å¹¾ç­†)
    #  end (int) -   çµæŸç¹ªè£½çš„è³‡æ–™åº(ç¬¬å¹¾ç­†)
    #  labels (list of strings)- å°æ–¼å¤šæ™‚é–“åºåˆ—æˆ–å¤šç¶­åº¦çš„æ¨™è¨»
    #  title (string)- åœ–ç‰‡æ¨™é¡Œ
    
    # è‹¥è³‡æ–™åªæœ‰ä¸€ç­†ï¼Œå‰‡è½‰ç‚ºlist
    if type(signal)!=list:
        signal=[signal]
        
    if not end:
        end=len(signal[0])
    time=(np.arange(len(signal[0]))/sr)[start:end]
    if labels:
    # è¨­ç«‹dictionary, è®“plotlyç•«è¨Šè™Ÿç·šæ™‚å¯ä»¥æ¨™è¨»label    
        dictionary={"time":time}
        for idx,l in enumerate(labels):
            # æˆªæ–·è³‡æ–™ï¼Œä¿ç•™æƒ³çœ‹çš„éƒ¨åˆ†ï¼Œä¸¦åˆ†æ®µç´€éŒ„æ–¼dictionaryä¸­
            dictionary.update({l:signal[idx][start:end]})
        # ç•«è¨Šè™Ÿç·š
        fig = px.line(dictionary,x="time",y=list(dictionary.keys())[1:],width=1000, height=400,title=title)
    else:
        # ç•«è¨Šè™Ÿç·š
        fig = px.line(x=time,y=signal,width=1000, height=400,title=title)
    fig.show()
````

## Fast Fourier Transform

åˆ†æè¨Šè™Ÿçš„æˆåˆ†æœ€å¸¸è¦‹çš„æ–¹å¼æ˜¯é »è­œåˆ†æï¼Œé€£çºŒè¨Šè™Ÿå¯ä»¥è—‰ç”±Fourier Transformå¾—åˆ°é »è­œï¼Œä¹Ÿå°±æ˜¯è¨Šè™Ÿåœ¨é »ç‡ä¸Šçš„åˆ†å¸ƒã€‚

è€Œå°æ–¼æ•¸ä½è¨Šè™Ÿï¼Œæˆ‘å€‘å¯ä»¥è—‰ç”±Discrete Fourier Transformå¾—åˆ°é »è­œ
$$ğ—_kâ‰”\sum\limits_{ğ‘›=0}^{ğ‘âˆ’1}ğ‘¥_ğ‘› ğ‘’^{(âˆ’ğ‘—2ğœ‹ğ‘˜ğ‘›/ğ‘)}$$

<img src=https://i.imgur.com/CZh8cYU.png width=800>


Scipyæˆ–è€…å…¶ä»–å¥—ä»¶åŒ…æœ‰æä¾›ä¸€äº›æ–¹å¼åšå¿«é€Ÿçš„Discrete Fourier Transformï¼Œç¨±ç‚ºFast Fourier Transform(FFT)ã€‚

è‹¥æ˜¯è¨Šè™Ÿåˆ†æå‰‡ç”¨Scipyå°±å¥½ï¼Œè‹¥æ˜¯è¦æ•´åˆåˆ°ç¥ç¶“ç¶²è·¯ä¸Šå¯èƒ½å°±å¾—ä½¿ç”¨Tensorflowæˆ–Pytorchå…§å»ºçš„fftã€‚

````python
from scipy import fft 
````

$$$$é€™é‚Šå› ç‚ºæˆ‘å€‘åšçš„æ˜¯é›¢æ•£çš„Fourierè½‰æ›ï¼Œè‹¥è¦çœ‹åˆ°åŸæœ¬å–®ä½(seconds, Hz)å‰‡éœ€é€²è¡Œè½‰æ›ã€‚
* æ™‚é–“: $n=f_s t$
* é »ç‡: $k=f N /f_s$

Nç‚ºåƒèˆ‡FFTçš„é»æ•¸ï¼Œ$f_s$å‰‡æ˜¯sampling rateã€‚

````python
## ä½¿ç”¨scipy.fft.fftå¯ä»¥å°è¨Šè™Ÿåšfft
# 'x' - è³‡æ–™
# 'n' - FFTé»æ•¸ï¼Œé€šå¸¸ä¸æŒ‡å®šï¼Œé è¨­ç‚ºè³‡æ–™ç¸½é»æ•¸
# è¨˜å¾—åšå®ŒFourier Transformå¾Œä¸ç®¡æ˜¯continuous/discrete/fast Fourier Transform, å‡ºä¾†éƒ½æ˜¯è¤‡æ•¸ (é è¨­ç‚ºcomplex64æ ¼å¼)
N=len(googl_)
googl_f = fft.fft(googl_[:N])
frequency=np.linspace(0,sr,N)
max_f=sr/2 # è¨­å®šæƒ³çœ‹åˆ°çš„æœ€å¤§é »ç‡ (Hz)
max_k=int(max_f*N/sr) # è½‰æˆk
plt.plot(frequency[:max_k], abs(googl_f[:max_k])) # é€šå¸¸æˆ‘å€‘æ˜¯çœ‹é€™å€‹è¤‡æ•¸çš„magnitude, å–absoluteå°±å¯ä»¥åšåˆ°
plt.xlabel("frequency(Hz)")
````

åœ¨é »è­œåœ–ä¸­æˆ‘å€‘å¯ä»¥çœ‹åˆ°è¼ƒç‚ºé«˜å³°çš„é»å°±è¡¨ç¤ºè¨Šè™Ÿæœ‰è¼ƒå¤šè©²é »ç‡æˆåˆ†ã€‚

æˆ‘å€‘å¯ä»¥è©¦è‘—æŠŠåšé »è­œçš„æ™‚é–“ç¸®é»ä¸€é»ï¼Œä¾†çœ‹çœ‹æœ€å‰é¢ä¸€æ®µçš„é »è­œ

````python
N1=int(0.12*sr)
N2=int(0.15*sr)
plt.plot(np.arange(N1,N2)/sr,googl_[N1:N2])
plt.xlabel("time(s)")
plt.show()
googl_f2 = fft.fft(googl_[N1:N2])
frequency2=np.linspace(0,sr,N2-N1)
max_f=4000 # è¨­å®šæƒ³çœ‹åˆ°çš„æœ€å¤§é »ç‡ (Hz)
max_k2=int(max_f*(N2-N1)/sr) # è½‰æˆk
plt.plot(frequency2[:max_k2],abs(googl_f2[:max_k2])) 
plt.xlabel("frequency(Hz)")
````

æˆ‘å€‘å–ç¬¬ä¸€å€‹éŸ³çš„é »è­œä¾†çœ‹å¯çœ‹åˆ°ä¸€å€‹ä¸»é »ç‡ï¼Œå°±èªéŸ³ä¾†è¬›æœƒæ¯”è¼ƒæ¨¡ç³Šä¸€é»

````python
plt.plot(frequency[:max_k],abs(googl_f[:max_k]))
plt.plot(frequency2[:max_k2],abs(googl_f2[:max_k2])) 
plt.xlabel("frequency(Hz)")
````

æ“·å–èˆ‡æ•´æ®µå°æ¯”:
* æ•´æ®µæ™‚é•·è¼ƒé•·ï¼Œå †ç©å‡ºçš„èƒ½é‡æ¯”è¼ƒé«˜
* æ•´æ®µéŸ³èª¿æ··é›œï¼Œå½¼æ­¤äº¤éŒ¯

äº¦å¯ç”¨```librosa.power_to_db```è½‰æ›æˆåˆ†è²é »è­œä¾†çœ‹ï¼Œæœƒä½¿scaleå·®æ¯”è¼ƒå¤šçš„éƒ¨åˆ†è¼ƒç‚ºç·©å’Œï¼Œæ–¹ä¾¿æ¯”è¼ƒ

````python
librosa.power_to_db(abs(googl_f2))
plt.plot(frequency[:max_k], librosa.power_to_db(abs(googl_f[:max_k])))
plt.plot(frequency2[:max_k2], librosa.power_to_db(abs(googl_f2[:max_k2]))) 
plt.xlabel("frequency(Hz)")
````

## Short-Time Fourier Transform

è‹¥æ˜¯æƒ³çµ±è¨ˆå‡ºä¸€æ®µå…§å‡ºç¾å“ªäº›éŸ³ç¬¦ï¼Œå‰‡é€™å€‹FFTå°±å¤ ç”¨äº†ï¼Œä½†é€šå¸¸æƒ³çŸ¥é“çš„æ˜¯åœ¨æ¯å€‹æ™‚é–“é»é »ç‡é•·ä»€éº¼æ¨£å­å°±å¿…é ˆè¦åšæ™‚é »åˆ†æã€‚

æ™‚é »åˆ†æä¸­æœ€å¸¸è¦‹çš„æ–¹å¼å°±æ˜¯Sort-Time Fourier Transform:
1. Windowing
2. å€‹åˆ¥åšé »è­œ

å…¬å¼:

$ğ‘‹[q,k]=\sum\limits_{ğ‘›^â€²=âŒˆâˆ’ğ‘/2âŒ‰}^{âŒˆğ‘/2âŒ‰âˆ’1}ğ‘¥[ğ‘›^â€²+ğ‘ğ»]ğ‘¤[ğ‘›^â€²] ğ‘’^\frac{âˆ’ğ‘—2ğœ‹ğ‘˜ğ‘›^â€²}{ğ‘}$

* q- æ¯å€‹windowçš„é›¢æ•£æ™‚é–“é»
* k- é›¢æ•£é »ç‡é»
* n'- åŸè¨Šè™Ÿçš„é›¢æ•£æ™‚é–“é»
* N- windowå…§FFTé»æ•¸ï¼ŒåŒæ™‚æ˜¯window size
* H- hop sizeï¼Œæ¯å€‹windowé›¢å¤šå°‘n'
* w- windowing function
* x- signal

<img src=https://i.imgur.com/Ji4xT6o.png width=600>


Librosaè£¡é¢æ²’æœ‰FFTä½†æœ‰å¾ˆå¤šæ™‚é »åˆ†æçš„å·¥å…·ã€‚

æˆ‘å€‘é€™é‚Šä½¿ç”¨```librosa.stft```ä¾†åšæ™‚é »åˆ†æ

````python
## ä½¿ç”¨librosa.stftå¯ä»¥å°è¨Šè™Ÿåšstft
# 'y' - è³‡æ–™
# 'n_fft' - FFTé»æ•¸ï¼ŒåŒæ™‚æ˜¯windowé•·åº¦ï¼Œé€™é‚Šå°±ä¸€å®šè¦æŒ‡å®šäº†ï¼Œé è¨­2048ï¼Œå¯æ ¹æ“šè‡ªå·±æƒ³çœ‹åˆ°çš„é »ç‡ç¯„åœèª¿æ•´
# 'hop_length' - æ¯å€‹windowé–“è¦è·³å¤šé•·
# 'window' - windowing functionï¼Œé è¨­æ˜¯ 'hann'
# å‡ºä¾†æ˜¯è¤‡æ•¸ (é è¨­ç‚ºcomplex64æ ¼å¼)

N=2048
H=512
googl_S = librosa.stft(googl_,n_fft=N, hop_length=H)
````

````python
# å¯ä»¥çœ‹å‡ºï¼Œæˆ‘å€‘ä½¿ç”¨librosa STFTå¯ç”Ÿå‡ºä¸€å€‹K x Q çš„çŸ©é™£ 
print(googl_S.shape)
print(type(googl_S))
print(type(googl_S[0,0]))
````

````python
# Kçš„å¤§å°ä¾†è‡ªN points/2ï¼Œå› ç‚ºå¤§æ–¼sampling rate/2çš„è¨Šè™Ÿæœƒaliase
print(N//2+1)
# Qçš„å¤§å°ä¾†è‡ªç¸½é•·åº¦é™¤ä»¥hop lengthï¼Œå°±æ˜¯windowingçš„æ¬¡æ•¸
print(len(googl_)//H+1)
````

### Spectral-centroid

è‹¥æƒ³è¦çŸ¥é“é€™å€‹Spectrogramä¸­é »ç‡å¤§è‡´ä¸Šåœ¨å“ªé‚Šä»¥åŠé »ç‡çš„è®ŠåŒ–å¯ä»¥è¨ˆç®—spectral centroid

$SC[q]=\sum\limits_{k=1}^N{\frac{F[q,k]}{\sum_{k'=1}^N{F[q,k']}}\times k}$

librosaæä¾›```librosa.feature.spectral_centroid```é€™å€‹åŠŸèƒ½

````python
## ä½¿ç”¨librosa.feature.spectral_centroidè¨ˆç®—é »ç‡ä¸­å¿ƒéš¨æ™‚é–“çš„è®ŠåŒ–
# 'y' or 'S' - å¯ä»¥é¸æ“‡ç”±è³‡æ–™yå¾è¨ˆç®—Spectrogramé–‹å§‹ï¼Œä¹Ÿå¯ä»¥ç›´æ¥ä¸ŸSpectrogram S çš„magnitude
# 'sr' - sampling rate
# è‹¥ä½¿ç”¨yç•¶inputæœƒæœ‰ä»¥ä¸‹åšSTFTçš„argumentå¯ä»¥ä½¿ç”¨:
## 'n_fft' - FFTé»æ•¸ï¼ŒåŒæ™‚æ˜¯windowé•·åº¦ï¼Œé€™é‚Šå°±ä¸€å®šè¦æŒ‡å®šäº†ï¼Œé è¨­2048ï¼Œå¯æ ¹æ“šè‡ªå·±æƒ³çœ‹åˆ°çš„é »ç‡ç¯„åœèª¿æ•´
## 'hop_length' - æ¯å€‹windowé–“è¦è·³å¤šé•·
## 'window' - windowing functionï¼Œé è¨­æ˜¯ 'hann'
# å›å‚³å–®ä½æ˜¯é »ç‡ Hz
mag_S=abs(googl_S)
googl_sc = librosa.feature.spectral_centroid(S=mag_S,sr=sr)
# N=2048
# H=512
# googl_sc = librosa.feature.spectral_centroid(y=googl_, sr=sr, n_fft=N , hop_length=H, window='hann' ) # ä¹Ÿå¯ä»¥å¾è¨Šè™Ÿé–‹å§‹
````

````python
# é€™é‚Šè©¦ä¸€å€‹Windowæ¯”è¼ƒå¤§çš„æƒ…æ³ï¼Œæœƒä½¿å¾—é »ç‡çš„è§£æåº¦è¼ƒé«˜ï¼Œä½†æ™‚é–“è§£æåº¦æœƒè¼ƒå·®
N=8192
H=512

googl_S = librosa.stft(googl_,n_fft=N, hop_length=H)
mag_S=abs(googl_S)
googl_sc = librosa.feature.spectral_centroid(S=mag_S,sr=sr)
googl_S_db = librosa.power_to_db(abs(googl_S)) # **
times = librosa.times_like(googl_sc)
fig, ax = plt.subplots()
ldp.specshow(googl_S_db, sr=sr, x_axis="s", y_axis="log",cmap='jet')
ax.plot(times, googl_sc.T, label='Spectral centroid', color='k')
ax.legend(loc='upper right')
````

## Mel-Spectrogram

åœ¨è²éŸ³è¨Šè™Ÿä¸Šï¼Œç‚ºæ¨¡æ“¬äººé¡è½è¦ºï¼Œå¸¸ä½¿ç”¨æ¨¡æ“¬äººè€³å°é »ç‡è®ŠåŒ–æ•æ„Ÿåº¦çš„Mel-Spectrogramä¾†åšæ™‚é »åˆ†æã€‚

å…¶ä¸­æœƒç”¨åˆ°Mel-scaleï¼Œæ˜¯Herzç¶“élogè½‰æ›å¾Œçš„å–®ä½ï¼Œæ˜¯å› æ‡‰é »ç‡è®ŠåŒ–æ•æ„Ÿåº¦exponentialä¸Šå‡çš„èª¿æ•´ã€‚

<img src='https://i.imgur.com/jgSQKv4.png' width=800>

åœ¨åšå®ŒSTFTå¾Œï¼Œå°K'å€‹ç­‰é•·é »æ®µ (åœ¨Mel-Scaleä¸Šç­‰é•·) é€²è¡Œtriangle filterå°±æ˜¯Mel-Spectrogramã€‚

ä½¿ç”¨çš„filter bankå¦‚ä¸‹åœ–æ‰€ç¤ºã€‚

e.g. 10 bins Mel filter bank for 65~1K Hz

<img src='https://i.imgur.com/xDgS9qj.png' width=600>

ä½¿ç”¨```librosa.feature.melspectrogram```å¯ä»¥é‡å°è¨Šè™Ÿè¨ˆç®—Mel Spectrogramã€‚

````python
## ä½¿ç”¨librosa.feature.melspectrogramè¨ˆç®—Mel-Spectrogram
# 'y' or 'S' - å¯ä»¥é¸æ“‡ç”±è³‡æ–™yå¾è¨ˆç®—Spectrogramé–‹å§‹ï¼Œä¹Ÿå¯ä»¥ç›´æ¥ä¸ŸSpectrogram S çš„magnitude
# 'sr' - sampling rate
# 'n_mels' - è¦æœ‰å¤šå°‘å€‹binï¼Œå°±æ˜¯å‰è¿°çš„K'
# è‹¥ä½¿ç”¨yç•¶inputæœƒæœ‰ä»¥ä¸‹åšSTFTçš„argumentå¯ä»¥ä½¿ç”¨:
## 'n_fft' - FFTé»æ•¸ï¼ŒåŒæ™‚æ˜¯windowé•·åº¦ï¼Œé€™é‚Šå°±ä¸€å®šè¦æŒ‡å®šäº†ï¼Œé è¨­2048ï¼Œå¯æ ¹æ“šè‡ªå·±æƒ³çœ‹åˆ°çš„é »ç‡ç¯„åœèª¿æ•´
## 'hop_length' - æ¯å€‹windowé–“è¦è·³å¤šé•·
## 'window' - windowing functionï¼Œé è¨­æ˜¯ 'hann'
# å›å‚³Mel-Spectrumï¼Œæ˜¯å¯¦æ•¸çŸ©é™£ï¼Œé è¨­ç‚ºfloat32çš„æ ¼å¼
N=2048
H=512
K_=256 # é€™Mel-spectrumçš„binæ•¸ä¸€å®šæœƒæ¯”å‰é¢Nä¾†å¾—å°‘ï¼Œæ‰æœ‰è¾¦æ³•åšfilterï¼Œä¸ç„¶ä¸­é–“æœƒæœ‰å¾ˆå¤šé »æ®µæœ‰ç¼ºå€¼
googl_mel_S=librosa.feature.melspectrogram(googl_,sr=sr,n_mels=K_,n_fft=N,hop_length=H,window='hann')
````

````python
# å¯ä»¥çœ‹ä¸€äº›åŸºæœ¬æ€§è³ª
print(googl_mel_S.shape)
print(googl_mel_S.min(),googl_mel_S.max())
print(type(googl_mel_S[0,0]))
````

åœ¨ç•«spectrogramæ™‚è¨˜å¾—æŠŠyè»¸scalingæ›æˆmel (æŒ‡çš„æ˜¯inputå–®ä½)ï¼Œé€™æ¨£ä»–æ‰æœƒå¹«å¿™è½‰å›Hzï¼Œä¸¦ä¸”ä»¥Mel scalingæ’åˆ—é¡¯ç¤º

librosaè¨­å®šçš„mel scaleè·Ÿlog scaleæœƒæœ‰ç¨å¾®ä¸åŒï¼Œè¼ƒå°ˆæ³¨æ–¼è¼ƒé«˜é »çš„éƒ¨åˆ†ã€‚

````python
googl_mel_S_db=librosa.power_to_db(googl_mel_S)
plt.figure(figsize=(6,5))
ldp.specshow(googl_mel_S_db,sr=sr,x_axis="s",y_axis="mel",cmap="jet") ## **
plt.colorbar(format="%+4.f")
plt.show()
````
