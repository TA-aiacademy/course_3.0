Laplacian Filter (5x5)：
$\begin{bmatrix}
0 & 0 &-1 & 0 & 0 \\
0 & -1 & -2 & -1 & 0\\
-1 & -2 & 16 & -2 & -1\\
0 & -1 & -2 & -1 & 0\\
0 & 0 &-1 & 0 & 0
\end{bmatrix}$

```python
path = "car.png"
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_img(path):
  img = cv2.imread(path)[:,:,::-1]
  laplacian_filter = np.array([[0, 0, -1, 0, 0],
                 [0, -1, -2, -1, 0],
                 [-1, -2, 16, -2, -1],
                 [0, -1, -2, -1, 0],
                 [0, 0, -1, 0, 0]])
  edge = cv2.filter2D(img, -1, laplacian_filter)
  return edge
```

```python
import cv2
import numpy as np
quiz_img = cv2.imread("car_ans.png")[:,:,::-1]
assert np.array_equiv(load_img(path), quiz_img)
```
