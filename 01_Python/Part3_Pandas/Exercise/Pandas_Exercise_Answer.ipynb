## $\Large{Pandas\; 練習題}$

```python
import pandas as pd
import numpy as np
```

## 範例資料

```python
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
```

## Exercise 1
以上面提供的字典資料建立dataframe資料，並且命名為df

```python
df = pd.DataFrame(data)
```

```python
df
```

## Exercise 2
使用describe呈現df資料的基本資訊

```python
df.describe()
```

## Exercise 3
從df資料中挑選animal與priority兩個欄位的資料

```python
df[['animal', 'priority']]
```

## Exercise 4
使用.loc的方式選取df資料中index為3, 4, 8且欄位為animal與age的資料

```python
df.loc[[3, 4, 8], ['animal', 'age']]
```

## Exercise 5
選取df資料中age欄位非遺漏值的所有資料

```python
df[~df.age.isna()]
```

## Exercise 6
承上，在剔除掉age的遺漏資料後將df資料以age欄位由小到大排序

```python
df = df[~df.age.isna()]
df.sort_values('age')
```

## Exercise 7
找出df資料中age的最大值

```python
df.age.max()
```

## Exercise 8
使用groupby方法依據priority欄位分組並計算visits的平均數

```python
df.groupby('priority').visits.mean()
```

## Exercise 9
繪製df資料中animal欄位的長條圖

```python
df.animal.value_counts().plot.bar()
```

## Exercise 10
繪製df資料中age欄位的機率密度函數圖

```python
df.age.plot.kde()
```
